```python
not_shuffled = tf.constant([[10,7],[5,3],[3,5]])
tf.random.shuffle(not_shuffled)
```

```python
tf.random.shuffle(not_shuffled,seed=42) #this does change all time
```

```python
tf.random.set_seed(42) #this does not change for each run
tf.random.shuffle(not_shuffled) #operational level seed 
#if both operational level seed and global level seed are same then it will suffle in same order for all runs
#in above example operational seed is automatically set to 42
```

## create a tensor from the numpy array


```python
# it creates the np arrary from 1 to 24
import numpy as np
array_A = np.arange(1,25,dtype=np.int32)
array_A
```

X= tf.constant capital letter refers to matrix or scalar
y = tf.Variable small letter refers variable

```python
A = tf.constant(array_A)
B = tf.constant(array_A,shape=(6,4))
A,B 
```

```python
tf.size(A).numpy()
tf.size(B).numpy()
```

```python
A.ndim
```

adding new dimensions with two methods
## expand_dims(array,axis)
##  [...,tf.newaxis]

```python
rank2 = A[...,tf.newaxis]
rank2
```

```python
rank_2 = tf.expand_dims(A,axis=0) #array will be unchanged and with shape 1,24
rank_2 
```

```python
C = tf.constant([[1,2],[3,4],[5,6]]) 
print(C)
rank_21 = tf.expand_dims(C,axis=-1)# this splits inner array into new array thus changes the shape (3,2,1)
rank_21
```
